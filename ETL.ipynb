```python
df=pd.read_csv("JC-201901-citibike-tripdata.csv")
```

```python
df.head()
```

```python
df.count()
```

```python
df_=pd.concat([df_jan,df])
```

```python
df_.head()
```

```python
df_.count()
```

```python
df_.keys()
```

```python
df_["age"]=2019 - df_["birth year"]
```

```python
df_["age"]
```

```python
df_gender=[]
for i in df_["gender"]:
    if i ==0:
#         print(0)
        df_gender.append("unknown")
    elif i==1:
#         print(1)
        df_gender.append("male")
    else: 
#         print(2)
        df_gender.append("female")

df_gender

df_["gendertext"]=df_gender
    
```

```python
df_.count()
```

```python
df_=df_[['tripduration', 'starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
        'gender', 'age']]
```

```python
df_.head()
```

```python
df_.to_csv("201901-citibike-cleandata.csv")
```
